```python
from scipy.stats import qmc
import matplotlib.pyplot as plt
import numpy as np



# lhd = qmc.LatinHypercube(d=2, num_samples=10)
d = 100
n = 128

dis_diff = []

for i in range(1000):

    sobol = qmc.Sobol(d=d, scramble=True)
    sobol_samp = sobol.random(n=n)

    lhc = qmc.LatinHypercube(d=d)
    lhc_samp = lhc.random(n=n)
    # sample



    lhc_dis = qmc.discrepancy(lhc_samp)
    sobol_dis = qmc.discrepancy(sobol_samp)

    dis_diff.append(sobol_dis - lhc_dis)

plt.plot(dis_diff)

#     plt.scatter(sobol_samp[:,0], sobol_samp[:,1], label='s')
#     plt.scatter(lhc_samp[:,0], lhc_samp[:,1], label='l')
# plt.legend()
```
