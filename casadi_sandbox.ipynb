```python
import casadi
from numpy import pi
import numpy as np

opti = casadi.Opti()

sin2 = opti.variable()
cos2 = opti.variable()
sin3 = opti.variable()
cos3 = opti.variable()

L1, L2, L3, L4, theta1 = (1, 2.5, 2.5, 2, 5/4*pi)
sin1, cos1 = (np.sin(theta1), np.cos(theta1))

opti.minimize(  (L1*sin1 + L2*sin2 + L3*sin3)**2 + (L1*cos1 + L2*cos2 + L3*cos3 - L4)**2)
opti.subject_to( sin2**2 + cos2**2==1 )
opti.subject_to( sin3**2 + cos3**2==1 )
# opti.subject_to(       x+y>=1 )

        # L1*sin1 + L2*sin2 + L3*sin3,
        # L1*cos1 + L2*cos2 + L3*cos3 - L4, 
        # sin2**2 + cos2**2 - 1,
        # sin3**2 + cos3**2 - 1,    


# (L1*sin1 + L2*sin2 + L3*sin3)**2 + (L1*cos1 + L2*cos2 + L3*cos3 - L4)**2 + (sin2**2 + cos2**2 - 1)**2 + (sin3**2 + cos3**2 - 1)**2


opti.solver('ipopt')


sol = opti.solve()

print(sol.value(sin2), sol.value(cos2), sol.value(sin3), sol.value(cos3))

sol.value((L1*sin1 + L2*sin2 + L3*sin3)**2 + (L1*cos1 + L2*cos2 + L3*cos3 - L4)**2)
```
