```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib widget
```

```python
df_all = pd.read_csv('cycle_data.csv')

df = df_all[df_all['t'] > 0.93]

# plt.figure()
plt.figure(figsize=(8,3))
plt.gca().set_aspect('equal', adjustable='box')


# plt.plot(df_all['x'], df_all['y'])
plt.plot(df['x'], df['y'], '-', color='magenta', linewidth=4)
plt.scatter(0,0,c='r')
plt.scatter(0.1533,0,c='r')

plt.axis('off')
plt.savefig('realtraj.png', transparent=True, bbox_inches='tight', pad_inches=0)

plt.figure(figsize=(8, 3))
plt.plot(df_all['t'], df_all['x'], label='x vs t')
plt.plot(df['t'], df['x'], '--', label='x vs t')

plt.plot(df_all['t'], df_all['y'], label='y vs t')
plt.plot(df['t'], df['y'], '--', label='y vs t')
plt.legend()
```
