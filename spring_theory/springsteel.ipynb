```python
from math import *

E_cf = 20 #msi
E_s = 29.7 

mu_s = 116 #kpsi
mu_cf = 120 #can be up to 175 kpsi

b_cf = 3 #mm
b_s = 3 #mm

h_cf = 0.5 #mm

h_s = h_cf * ((E_cf/E_s)*(b_cf/b_s))**(1/3)


import matplotlib.pyplot as plt

mu_cf_values = range(120, 176, 5)
h_s_values = []
h_s_yield_values = []

for mu_cf in mu_cf_values:
    h_s = h_cf * ((E_cf/E_s)*(b_cf/b_s))**(1/3)
    h_s_yield = h_cf*(mu_s/mu_cf)*(E_s/E_cf)
    h_s_values.append(h_s)
    h_s_yield_values.append(h_s_yield)

plt.plot(mu_cf_values, h_s_values, label=f'for same stiffness {h_s}')


plt.plot(mu_cf_values, h_s_yield_values, label='safe region')
plt.fill_between(mu_cf_values, h_s_values, h_s_yield_values, where=[h_s < h_s_yield for h_s, h_s_yield in zip(h_s_values, h_s_yield_values)], color='red', alpha=0.1)


plt.xlabel('mu_cf (kpsi)')
plt.ylabel('h_s (mm)')
plt.title('h_s vs mu_cf')
plt.grid(True)
plt.legend()
plt.show()
```
