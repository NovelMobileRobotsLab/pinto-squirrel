```python
from casadi import *
import numpy as np

X0 = SX.sym('X0', 3)
T = SX.sym('T', 2)
X_new = SX.sym('X_new', 3)
T1 = T[0]
T2 = T[1]
x_a0 = X0[0]
x_l0 = X0[1]
v_l0 = X0[2]
#calcs
x_a = x_a0 + T1
x_l = x_l0 + T2
v_l = v_l0 + T1 + T2
X_new[0] = x_a
X_new[1] = x_l
X_new[2] = v_l

step = Function("step", [X0, T], [X_new])

opti = Opti()
Ts = opti.variable(2, 3) #column vectors of [T1, T2], concat horizontally nsteps number of times

X_init = (0,0,0)
# X1 = step(X_init, Ts[0,:])
# X2 = step(X1, Ts[1,:])
# X3 = step(X2, Ts[2,:])

fX3_fold = step.fold(3)
X3_fold = fX3_fold(X_init, Ts)

F = X3_fold[2]

opti.minimize(-F)
opti.subject_to(opti.bounded(0, Ts[0,:], 2))
opti.subject_to(opti.bounded(0, Ts[1,:], 3))
opti.set_initial(Ts, 1)

log = []
opti.callback(lambda i: log.append(opti.debug.value(Ts)))
opti.solver('ipopt', {'ipopt.print_level':0, 'print_time':0})
sol = opti.solve()

print(sol.value(Ts), sol.value(F))
print(np.array(log))
```
